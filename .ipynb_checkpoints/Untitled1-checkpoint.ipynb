```python
import torch
import torchvision
import torchvision.transforms as transforms
import os
import torchvision.datasets as datasets
import torch.utils.data as data
import torch
from torch import nn
from torch import optim
import torch.nn.functional as F
from torchvision import models
```

```python
data_dir = 'C:/Users/nanda/Desktop/My_projects/flower_data'
train_dir = data_dir + '/train'
valid_dir = data_dir + '/valid'
file_dir = os.listdir(train_dir)
```

```python
# Define your transforms for the training, validation, and testing sets
data_transforms = {
    'train': transforms.Compose([
        transforms.RandomRotation(45),
        transforms.Resize(224),
        transforms.CenterCrop(224),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
     ]),
    'valid': transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
     ]),
}
    
# Load the datasets with ImageFolder
image_datasets = {
    x: datasets.ImageFolder(root=data_dir + '/' + x, transform=data_transforms[x])
    for x in list(data_transforms.keys())
}

dataloaders = {
    x: data.DataLoader(image_datasets[x], batch_size=4, shuffle=True, num_workers=2)
    for x in list(image_datasets.keys())
}
```

```python
import json

with open('cat_to_name.json', 'r') as f:
    cat_to_name = json.load(f)
```

```python
import pandas as pd 
import numpy as np


# converting json dataset from dictionary to dataframe
train_labels = pd.DataFrame.from_dict(cat_to_name, orient='index')
#train_labels.reset_index(level=0, inplace=True)
train_labels['file_number'] = train_labels.index
train_labels.index = range(len(train_labels))
```

```python
labs= np.array(train_labels[0])
labs[0]
```

```python
import matplotlib.pyplot as plt
import numpy as np

# functions to show an image


def imshow(img):
    img = img / 2 + 0.5     # unnormalize
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.show()


# get some random training images

inputs, classes = next(iter(dataloaders['train']))

# show images
imshow(torchvision.utils.make_grid(inputs))
# print labels
print(' '.join('%5s' % labs[classes[j]] for j in range(4)))
```

```python
import torchvision.models as model
VGG_19 = model.vgg19(pretrained=True)
for param in VGG_19.parameters():
    param.requires_grad = False
print(VGG_19)
```

```python

from collections import OrderedDict
classifier = nn.Sequential(OrderedDict([
                          ('fc1', nn.Linear(1024, 500)),
                          ('relu', nn.ReLU()),
                          ('fc2', nn.Linear(500, 102)),
                          ('output', nn.LogSoftmax(dim=1))
                          ]))
    
model.classifier = classifier
```

```python
criterion = nn.NLLLoss()
    # Only train the classifier parameters, feature parameters are frozen
optimizer = optim.Adam(model.classifier.parameters(), lr=0.001)
outputs = model.forward(inputs)
loss = criterion(outputs, clases)
loss.backward()
optimizer.step()

   
```
